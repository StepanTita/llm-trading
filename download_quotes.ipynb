````python
def get_url(exchange, symbol, year, month):
    return f'https://datasets.tardis.dev/v1/{exchange}/trades/{year}/{month}/01/{symbol}.csv.gz'
````

````python
def get_exchange_details(exchange):
    response = requests.get(f'https://api.tardis.dev/v1/exchanges/{exchange}')
    return response.json()
````

````python
def get_available_dates(exchange, symbol):
    details = get_exchange_details(exchange)['datasets']['symbols']
    items = [d['id'] for d in details]
    since = datetime.strptime(details[items.index(symbol)]['availableSince'],
                              "%Y-%m-%dT%H:%M:%S.%fZ")
    until = datetime.strptime(details[items.index(symbol)]['availableTo'],
                              "%Y-%m-%dT%H:%M:%S.%fZ")
    return since, until
````

````python
urls = []
for exchange in exchanges:
    for symbol in symbols[exchange]:
        since, until = get_available_dates(exchange, symbol)

        for year in range(since.year, until.year + 1):
            if year == until.year:
                months = [f'{i:02d}' for i in range(since.month + 1, until.month + 1)]
            elif year == since.year:
                months = [f'{i:02d}' for i in range(since.month + 1, 13)]
            else:
                months = [f'{i:02d}' for i in range(1, 13)]

            for month in months:
                urls.append((get_url(exchange, symbol, year, month), symbol, f'{year}_{month}'))
urls
````

````python
def download_file(url, filename):
    try:
        response = requests.get(url)

        # check if response is json
        if response.headers['Content-Type'] == 'application/json':
            print(f'Failed to download {url}')
            print('Error:', response.json())
            return False

        with open(filename, 'wb') as file:
            file.write(response.content)
            print(f'Saved {filename}')
    except Exception as e:
        print(f'Failed to download {url}')
        print('Error:', e)
    return True
````
